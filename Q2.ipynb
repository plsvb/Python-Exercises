```python
import pandas as pd


url = 'https://raw.githubusercontent.com/WHPAN0108/BHT-DataScience-S23/main/python-DS/country.csv'
df = pd.read_csv(url)
print("Daten erfolgreich geladen!")
print()

print("Deskriptive Statistiken für numerische Spalten:")
print(df.describe())
print()

print("Die letzten 4 Zeilen des DataFrame:")
print(df.tail(4))
print()

print("Länder mit der Währung EURO:")
euro_countries = df[df['Currency'] == 'EUR']
print(euro_countries)
print()

print("DataFrame nur mit Name und Währung:")
name_currency_df = df[['Name', 'Currency']]
print(name_currency_df)
print()

print("Länder mit einem GDP über 2000 Milliarden USD:")
high_gdp_countries = df[df['GDP'] > 2000]
print(high_gdp_countries)
print()

print("Länder mit einer Einwohnerzahl zwischen 50 und 150 Millionen:")
mid_pop_countries = df[(df['Population'] > 50000000) & (df['Population'] < 150000000)]
print(mid_pop_countries)
print()

print("Durchschnittliches GDP (fehlende Werte ignorieren):")
average_gdp = df['GDP'].mean(skipna=True)
print(average_gdp)
print()

print("Durchschnittliches GDP (fehlende Werte als 0 behandelt):")
df_filled = df.fillna({'GDP': 0})
average_gdp_zero = df_filled['GDP'].mean()
print(average_gdp_zero)
print()

print("Bevölkerungsdichte:")
df['Population Density'] = df['Population'] / df['Area']
print(df[['Name', 'Population Density']])
print()

print("Nach Ländernamen alphabetisch sortiert:")
df_sorted = df.sort_values(by='Name')
print(df_sorted)
print()








```
